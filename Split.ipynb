```python
import pandas as pd
from sklearn.model_selection import train_test_split
```

```python
data = pd.read_csv('appointment_data.csv')


data
```

```python
y= data['No-show']
y
```

```python
X= data.drop('No-show',axis=1)
X
```

```python
#Train/Test Split
train_X, test_X, train_y, test_y = train_test_split(X, y, test_size = 0.2)
```

```python
train_X.insert(0,train_y.name,train_y)
train_X
```

```python
test_X.insert(0,test_y.name, test_y)
test_X
```

```python
train_X.to_csv('train.csv',index=0)
```

```python
test_X.to_csv('test.csv',index=0)
```
